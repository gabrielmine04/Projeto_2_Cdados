```python
def regress(Y,X):
    '''
    Y: coluna do DataFrame utilizada como variÃ¡vel resposta (TARGET)
    X: coluna(s) do DataFrame utilizadas como variÃ¡vel(is) explicativas (FEATURES)
    '''
    X_cp = sm.add_constant(X)
    model = sm.OLS(Y,X_cp)
    results = model.fit()
    
    return results
```

```python
X = w[['gdp', 'literacy', 'eletricity', 'dolar',
       'acess', 'celular']]
Y = w['net']
results = regress(Y,X)
results.summary()
```
